**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Step 2: Importing dataset**

```python
data = pd.read_csv('/content/Data.csv')
data
```

```python
data.head()
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data.dtypes
```

**Step 3: Handling the missing data**

```python


data.isnull().sum()
```

```python
data
```

```python
data['Age'] =data['Age'].interpolate(method = 'linear',limit_direction = 'forward')
data
```

```python
data['Salary'] =data['Salary'].interpolate(method = 'linear',limit_direction = 'forward')
data
```

**Step 4: Encoding categorical data**

```python
data['Country'].value_counts()
```

```python
data['Purchased'].value_counts()
```

```python
data['Purchased'] = data['Purchased'].map({'Yes':0,'No':1}).values
```

```python
data.dtypes
```

**Step 5: Creating a dummy variable**

```python
data = pd.get_dummies(data,columns =['Country'],drop_first=True)
```

```python
data
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X = data.drop(['Purchased'],axis = 1).values
y = data['Purchased'].values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size =0.2)
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

```
